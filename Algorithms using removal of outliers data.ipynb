```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(15, 10))
sns.boxplot(x = 'MonthlyIncome', y = 'Gender', data = employee_df)
```

```python
employee_df1= employee_df[employee_df.MonthlyIncome<=(employee_df.MonthlyIncome.mean() + 1*employee_df.MonthlyIncome.std())]
employee_df1
```

```python
plt.figure(figsize=(15, 10))
sns.boxplot(x = 'MonthlyIncome', y = 'Gender', data = employee_df1)
```

```python
employee_df2= employee_df1[employee_df1.TotalWorkingYears<=(employee_df1.TotalWorkingYears.mean() + 1*employee_df1.TotalWorkingYears.std())]
employee_df2
```

```python
employee_df3= employee_df2[employee_df2.TrainingTimesLastYear<=(employee_df2.TrainingTimesLastYear.mean() + 1*employee_df2.TrainingTimesLastYear.std())]
employee_df3
```

```python
employee_df4= employee_df3[employee_df3.TotalWorkingYears<=(employee_df3.TotalWorkingYears.mean() + 1*employee_df3.TotalWorkingYears.std())]
employee_df4
```

```python
employee_df5= employee_df4[employee_df4.YearsAtCompany<=(employee_df4.YearsAtCompany.mean() + 1*employee_df4.YearsAtCompany.std())]
employee_df5
```

```python
employee_df6= employee_df5[employee_df5.YearsInCurrentRole<=(employee_df5.YearsInCurrentRole.mean() + 1*employee_df5.YearsInCurrentRole.std())]
employee_df6
```

```python
employee_df7= employee_df6[employee_df6.YearsWithCurrManager<=(employee_df6.YearsWithCurrManager.mean() + 1*employee_df6.YearsWithCurrManager.std())]
employee_df7
```

```python
y = employee_df7['Attrition']
y.value_counts()
```

```python
sns.countplot(employee_df7['Attrition'])
```

```python
from sklearn.preprocessing import StandardScaler, LabelEncoder
for column in employee_df7.columns:
    if employee_df7[column].dtype == np.number:
        continue
    employee_df7[column] = LabelEncoder().fit_transform(employee_df7[column])
    
employee_df7
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
X = scaler.fit_transform(employee_df7.drop('Attrition',axis = 'columns'))
X
```

```python
y = employee_df7['Attrition']
y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25)
X_train.shape
```

```python
X_test.shape
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

model = LogisticRegression()
model.fit(X_train, y_train)

y_pred = model.predict(X_test)
y_pred
```

```python
print("Accuracy {} %".format( 100 * accuracy_score(y_pred, y_test)))
```

```python
from sklearn.metrics import confusion_matrix, classification_report
print(classification_report(y_test, y_pred))
```

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
model = DecisionTreeRegressor()
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
y_pred
```

```python
print("Accuracy {} %".format( 100 * accuracy_score(y_pred, y_test)))
```

```python
cm = confusion_matrix(y_pred, y_test)
sns.heatmap(cm, annot=True)
```

```python
print(classification_report(y_test, y_pred))
```

```python
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier()
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
y_pred
```
